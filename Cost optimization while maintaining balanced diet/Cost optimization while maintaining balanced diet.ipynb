```python
##Libraries
from pulp import *
import pandas as pd
import numpy as np
```

```python
## Reading raw dataset
DietRaw = pd.read_excel('\diet.xls')
DietRaw.head()
```

```python
##Summary
print(DietRaw.shape)
print('-------------')
print('\n')
print(DietRaw.info())
print('-------------')
print('\n')
print(DietRaw.isnull().sum())
```

```python
##Remove rows with Nas
print(DietRaw.shape)
DietRaw = DietRaw.dropna()
print(DietRaw.shape)
```

```python
##List of all food items
FoodItems = list(DietRaw['Foods'])
FoodItems
```

```python
##Defining the minimization problem
DietProb = LpProblem("Optimize Diet",LpMinimize)
```

```python
## Creating a dictionary to store the food item with the respect price per serving
FoodCost = dict(zip(DietRaw['Foods'],DietRaw['Price/ Serving']))
FoodCost
```

```python
## A dictionary called 'food_vars' is created to contain the referenced Variables
## First argument is the name prefix, second are the reference variable, third is the lower bound and fourth is the type of values the variables can take
food_vars = LpVariable.dicts("Food",FoodItems,0,cat='Continuous')
food_vars
```

```python
## Objective funtion- Minimize cost
DietProb += lpSum([FoodCost[i] * food_vars[i] for i in FoodItems]), "Cost of Diet"
DietProb
```

```python

## Dictionary to store nutrient content per food item
calories = dict(zip(FoodItems,DietRaw['Calories']))
cholesterol = dict(zip(FoodItems,DietRaw['Cholesterol mg']))
fat = dict(zip(FoodItems,DietRaw['Total_Fat g']))
sodium = dict(zip(FoodItems,DietRaw['Sodium mg']))
carbs = dict(zip(FoodItems,DietRaw['Carbohydrates g']))
fiber = dict(zip(FoodItems,DietRaw['Dietary_Fiber g']))
protein = dict(zip(FoodItems,DietRaw['Protein g']))
vit_A = dict(zip(FoodItems,DietRaw['Vit_A IU']))
vit_C = dict(zip(FoodItems,DietRaw['Vit_C IU']))
calcium = dict(zip(FoodItems,DietRaw['Calcium mg']))
iron = dict(zip(FoodItems,DietRaw['Iron mg']))


##Adding in nutritional constraints

# Calorie
DietProb += lpSum([calories[f] * food_vars[f] for f in FoodItems]) <= 2500.0, "CalorieMax"
DietProb += lpSum([calories[f] * food_vars[f] for f in FoodItems]) >= 1500.0, "CalorieMin"

# Cholesterol
DietProb += lpSum([cholesterol[f] * food_vars[f] for f in FoodItems]) >= 30.0, "CholesterolMinimum"
DietProb += lpSum([cholesterol[f] * food_vars[f] for f in FoodItems]) <= 240.0, "CholesterolMaximum"

# Fat
DietProb += lpSum([fat[f] * food_vars[f] for f in FoodItems]) >= 20.0, "FatMinimum"
DietProb += lpSum([fat[f] * food_vars[f] for f in FoodItems]) <= 70.0, "FatMaximum"

# Sodium
DietProb += lpSum([sodium[f] * food_vars[f] for f in FoodItems]) >= 800.0, "SodiumMinimum"
DietProb += lpSum([sodium[f] * food_vars[f] for f in FoodItems]) <= 2000.0, "SodiumMaximum"

# Carbs
DietProb += lpSum([carbs[f] * food_vars[f] for f in FoodItems]) >= 130.0, "CarbsMinimum"
DietProb += lpSum([carbs[f] * food_vars[f] for f in FoodItems]) <= 450.0, "CarbsMaximum"

# Fiber
DietProb += lpSum([fiber[f] * food_vars[f] for f in FoodItems]) >= 125.0, "FiberMinimum"
DietProb += lpSum([fiber[f] * food_vars[f] for f in FoodItems]) <= 250.0, "FiberMaximum"

# Protein
DietProb += lpSum([protein[f] * food_vars[f] for f in FoodItems]) >= 60.0, "ProteinMinimum"
DietProb += lpSum([protein[f] * food_vars[f] for f in FoodItems]) <= 100.0, "ProteinMaximum"

# Vitamin A
DietProb += lpSum([vit_A[f] * food_vars[f] for f in FoodItems]) >= 1000.0, "VitaminAMinimum"
DietProb += lpSum([vit_A[f] * food_vars[f] for f in FoodItems]) <= 10000.0, "VitaminAMaximum"

# Vitamin C
DietProb += lpSum([vit_C[f] * food_vars[f] for f in FoodItems]) >= 400.0, "VitaminCMinimum"
DietProb += lpSum([vit_C[f] * food_vars[f] for f in FoodItems]) <= 5000.0, "VitaminCMaximum"

# Calcium
DietProb += lpSum([calcium[f] * food_vars[f] for f in FoodItems]) >= 700.0, "CalciumMinimum"
DietProb += lpSum([calcium[f] * food_vars[f] for f in FoodItems]) <= 1500.0, "CalciumMaximum"

# Iron
DietProb += lpSum([iron[f] * food_vars[f] for f in FoodItems]) >= 10.0, "IronMinimum"
DietProb += lpSum([iron[f] * food_vars[f] for f in FoodItems]) <= 40.0, "IronMaximum"
```

```python
##Running the optimizer
DietProb.solve()
```

```python

print("Therefore, the optimal (least cost) balanced diet consists of\n"+"-"*100)
for v in DietProb.variables():
    if v.varValue>0:
        print(v.name, "=", v.varValue)
```

```python
## Variable to define if a food item is chosen or not 1=chosen
food_vars2 = LpVariable.dicts("FoodSelected",FoodItems,cat='Binary')

## Constraint for setting food_var2 as 1 for chosen food items
for f in FoodItems:
    DietProb += (food_vars[f] * 0.00001) <= food_vars2[f], "Picked1"+f
    DietProb += food_vars[f] - food_vars2[f] >= -0.999999, "Picked2"+f


## Constraint to choose at 1/10 serving of each food item
for f in FoodItems:
    DietProb += food_vars[f] - food_vars2[f] >=-0.9, "Serving"+f

```

```python
## Constraint to choose either Frozen Broccolli or Celery but not both

##Dictionary to store the values
DietRaw['CeleryOrBroccoli'] = np.where((DietRaw['Foods']=='Frozen Broccoli') | (DietRaw['Foods']=='Celery, Raw'),1,0 )
CeleryOrBroccoli = dict(zip(FoodItems,DietRaw['CeleryOrBroccoli']))
CeleryOrBroccoli

##Constraint
DietProb += lpSum([CeleryOrBroccoli[f] * food_vars2[f] for f in FoodItems]) <= 1, "CeleryOrBroccoli"
```

```python
## Constraint to choose at least 3 kinds of meat/poultry/fish/eggs 

##List of items classified under meat/poultry/fish/eggs
MeatFoodItem = [
 'Roasted Chicken',
 'Poached Eggs',
 'Scrambled Eggs',
 'Bologna,Turkey',
 'Frankfurter, Beef',
 'Ham,Sliced,Extralean',
 'Kielbasa,Prk',
 'Pizza W/Pepperoni',
 'Hamburger W/Toppings',
 'Hotdog, Plain',
 'Pork',
 'Sardines in Oil',
 'White Tuna in Water',
 'Chicknoodl Soup',
 'Splt Pea&Hamsoup',
 'Vegetbeef Soup',
 'Beanbacn Soup,W/Watr']

##Dictionary to store the values
DietRaw['MeatItem'] = np.where(DietRaw['Foods'].isin(MeatFoodItem),1,0 )
MeatItem = dict(zip(FoodItems,DietRaw['MeatItem']))
MeatItem

##Constraint
DietProb += lpSum([MeatItem[f] * food_vars2[f] for f in FoodItems]) >= 3, "MeatItem"

```

```python
##Running the optimizer
DietProb.solve()
```

```python
##Solution
print("Optimal (least cost) balanced diet with additional constrainsts consists of\n"+"-"*100)
for v in DietProb.variables():
    if ((v.varValue>0) & ('Food_' in v.name)):
        print(v.name, "=", v.varValue)
```
